```python
import numpy as np
from scipy import stats
```

### Пункт а. Влиние возраста на содержание иммуноглобина

```python
Psi = np.zeros((25,5))
Psi[0:2, 0] = 1
Psi[2:8, 1] = 1
Psi[8:19, 2] = 1
Psi[19:23, 3] = 1
Psi[23:25, 4] = 1

Y = np.array([
    83, 85,
    84, 85, 85, 86, 86, 87,
    86, 87, 87, 87, 88, 88, 88, 88, 88, 89, 90,
    89, 90, 90, 91,
    90, 92
])

F = Psi.T @ Psi
F_inv = np.linalg.inv(F)
beta = F_inv @ Psi.T @ Y
print(beta)
```

```python
e = Y - Psi @ beta
rss = e @ e
print(rss)
```

### Значимость коэффициентов

```python
n, p = Psi.shape
for i in range(5):
    delta = (beta[i] * np.sqrt(n-p))/np.sqrt(rss*F_inv[i, i])
    p_value = 2 * stats.t.sf(delta, n-p)
    print(f"коэффициент {i+1}, p-value = {p_value}")
```

### Пункт б. Попарное сравнение средних с учётом множественности проверяемых гипотез

```python
for i in range(5):
    for j in range(5):
        if j < i or i==j:
            continue
        delta = (beta[i]-beta[j]) * np.sqrt((n-p)/(rss*(F_inv[i,i] + F_inv[j,j])))
        p_value = 2 * stats.t.sf(abs(delta), n-p)
        if p_value < 0.05:
            print(i + 1, j + 1," p-value =", p_value, "различимы")
        else:
            print(i + 1, j + 1," p-value =", p_value, "неразличимы")
```
