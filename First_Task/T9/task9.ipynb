### Расчёт дельты

```python
n1 = 300
n2 = 300

m = [
    [33, 43, 80, 144],
    [39, 35, 72, 154]
]
p = [72/600, 78/600, 152/600, 298/600]

delta1 = 0
delta2 = 0
for i in range(4):
    delta1 += ((m[0][i] - n1*p[i])**2)/(n1*p[i])
    delta2 += ((m[1][i] - n2*p[i])**2)/(n2*p[i])
delta = delta1 + delta2
print(delta)
```

### Расчёт p-value

```python
from scipy.stats import chi2

p_value = chi2.sf(delta, 3)
print(p_value)
```
