### Расчёт дельты (Меры расхождения)

```python
n = 200

p = [77/200, 91/200, 32/200]
q = [0.5, 0.5]


m = [
    [25, 50, 25],
    [52, 41, 7]
]

delta = 0
for i in range(2):
    for j in range(3):
        delta += ((m[i][j] - n*p[j]*q[i])**2)/(n*p[j]*q[i])
print(delta)
```

### Расчёт p-value

```python
from scipy.stats import chi2

p_value = chi2.sf(delta, 2)
print(p_value)
```
