```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.05
x = np.array([-1.11, -6.10, 2.42])
y = np.array([-2.29, -2.91])
sigma2_x = 2
sigma2_y = 1
```

### Параметры выборки

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

Sx = np.std(x)
Sy = np.std(y)
```

### Критическая область a>b

```python
c = norm.ppf(1-alpha)
print(c)
```

```python
sample_delta = np.sqrt(6/7)*(x_mean - y_mean)
print(sample_delta)
```

### Мощность

```python
def build_plot():
    abscissa = np.linspace(-3, 6, 10000)
    ordinate = 1 - norm.cdf(c - np.sqrt(6/7)*abscissa)

    figure, axis = plt.subplots()
    axis.plot(abscissa, ordinate, color = "cornflowerblue")
    axis.grid()
    axis.set_xlabel("d = a-b")
    axis.set_ylabel("W - мощность")
    plt.show()
build_plot()
```
