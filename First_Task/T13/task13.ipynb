### Константы и импорты

```python
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.05
n = 139
m = 1000
```

### Квантили распределения Фишера

```python
left_board = f.ppf(alpha/2, n-1, m-1 )
right_board = f.ppf(1-alpha/2, n-1, m-1)

print(left_board, right_board)
```

### Дельта для длины

```python
Sx_len = 5.722
Sy_len = 6.161
delta_len = (Sx_len**2)/(Sy_len**2)
print(delta_len)
```

### Дельта для ширины

```python
Sx_width = 4.612
Sy_width = 5.055
delta_width = (Sx_width**2)/(Sy_width**2)
print(delta_width)
```

### График мощности

```python
def build_plot():
    abscissa = np.linspace(0, 4, 1000)
    ordinate = f.cdf(left_board/abscissa,n-1,m-1) + 1-f.cdf(right_board/abscissa,n-1,m-1)

    figure, axis = plt.subplots()
    axis.plot(abscissa, ordinate, color = "cornflowerblue")
    axis.grid()
    axis.set_xlabel("c = a^2/b^2")
    axis.set_ylabel("W - мощность")
    plt.show()

build_plot()
```
